```python
# Import relevant dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV
def load_CSV(dataframe: pd.DataFrame)-> pd.DataFrame:
    """
    Load CSV to start process.
    """

    # Match column names with more appropriate ones
    column_matching = {
        'date': 'date',
        'admin1': 'region',
        'admin2': 'ISO',
        'market': 'city',
        'market_id': 'market_id',
        'latitude': 'latitude',
        'longitude': 'longitude',
        'category': 'crop_category',
        'commodity': 'crop',
        'commodity_id': 'crop_id',
        'unit': 'unit(KG)',
        'priceflag': 'price_flag',
        'pricetype': 'price_type',
        'currency': 'currency',
        'price': 'cedi_price',
        'usdprice': 'dollar_price'
    }
    
    df = pd.read_csv(dataframe, parse_dates=['date'], dayfirst=True)
    
    # Implement column matching
    df = df.rename(columns=column_matching)
    
    print('CSV Loaded.\n')
    
    return df
```

```python
# Drop tables
def drop_tables(df: pd.DataFrame = load_CSV('original_dataset/wfp_food_prices_gha.csv')) -> pd.DataFrame:
    """
    Drop specific columns.
    """
    df = df.drop(columns=['currency', 'dollar_price'], axis=1)
    
    print('Tables dropped.\n')
    
    return df
```

```python
# Extract dates
def date_engineering(df: pd.DataFrame = drop_tables()) -> pd.DataFrame:
    """
    Engineer year, month, and day from the data frame.
    """

    # Extract the year
    if 'year' not in df.columns:
        df.insert(loc=1, column=['year'], value=df['date'].dt.year, allow_duplicates=True)
    
    # Extract the month
    if 'month' not in df.columns:
        df.insert(loc=2, column=['month'], value=df['date'].dt.month, allow_duplicates=True)
    
    # Extract the day
    if 'day' not in df.columns:
        df.insert(loc=3, column=['day'], value=df['date'].dt.day, allow_duplicates=True)

    print('Years, months, and days engineered.\n')
    
    return df
```

```python
# Unique crops
def unique_crop(df: pd.DataFrame = date_engineering()) -> np.array:
    """
    Create an array of crops from the dataframe.
    """

    unique_crop = list(df['crop'].unique())

    print('Unique crops identified.\n')

    return unique_crop
```

```python
# Remove strings
def outlawed_words()-> list:
    """
    Generate strings to be removed from unit(KG) column.
    """
    outlawed_words = ['KG', ' KG', 'Tubers', ' Tubers', 'Bunch', ' Bunch', 'pcs', ' pcs']

    return outlawed_words
```

```python
# Split data frame
def split_dataframe(df: pd.DataFrame = date_engineering()) -> pd.DataFrame:
    """
    Split data frame into smaller data frames according to crops.
    """

    # Count number of crops    
    n= 0
    
    # Retrieve list of unique items
    for unique in unique_crop():
    

        # Retrieve original data frame and subset crop
        unique_df = df[df['crop'] == unique]
        unique_df = unique_df.drop(['crop_category', 'crop', 'crop_id'], axis=1)

        # Retrieve outlawed words and iterate
        for outlaw in outlawed_words():
            
            if unique_df['unit(KG)'].str.contains(outlaw, na=False).any():
                unique_df['unit(KG)'] = unique_df['unit(KG)'].str.replace(outlaw, '')
            else:
                continue
    
        unique = unique.replace(' ', '')

        # Save split data frame to folder
        unique_df.to_csv(f'crop_table/{unique}.csv', index=False)

        n += 1

    print('Crop-split CSVs created.\n')
    
    # Retrun number of split files
    return n
```

```python
# Main function
def main():
    """
    Main function to begin split process
    """
    number_of_splits = split_dataframe()
    
    print(f"{number_of_splits} data frames were created for {number_of_splits} crops.\n")
```

```python
main()
```
