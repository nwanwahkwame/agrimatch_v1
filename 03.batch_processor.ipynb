```python
import pandas as pd
from glob import glob
```

```python
# Find all CSV files in crop directory
crop_files = glob(f"crop_table/*.csv")
```

```python
# Process each crop dataset

def batch_processing(crop_files = crop_files):
    """
    Batch processes the CSV files in determined path.

    Parameters:
    crop_file = This is a glob instance that retrieves all files from a given path.
    """

    print(f'This is how I feel right now. 😒\n')
    
    for index, file in enumerate(crop_files, start=1):
        df = pd.read_csv(file)
        print(f"Processing File No. {index}: {file}...\n")
    
        # Drop price_type, price_flag, and market_id colums
        df.drop(['price_type', 'price_flag', 'market_id', 'market_sold_at'], axis=1, inplace=True)
        
        # Fill missing values with 1KG
        df['unit(KG)'] = df['unit(KG)'].fillna(1)
        
        # Standardize price per KG
        df['cedi_price/(KG)'] = round(df['cedi_price']/df['unit(KG)'], 2)

        # Drop unit(KG) after standardizing
        df.drop(['unit(KG)'], axis=1, inplace=True)
        
        # File name
        name = file.split('\\')[1]

        # Save to same path
        df.to_csv(path_or_buf=f'crop_table/{name}', index=False)

    print('Processing done. 😊')
```

```python
batch_processing()
```
