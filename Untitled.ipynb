```python
# Import Pandas
import pandas as pd
import os
from glob import glob
```

```python
region_files = glob(f'data/region_splits/*.csv')
```

```python
# Read CSV
df = pd.read_csv('data/original_dataset/wfp_food_prices_gha_clean.csv')
```

```python
# Set 1 as standard unit
df['unit'][df.loc[:, 'unit'] == '0'] = 1
```

```python
# Standard units to remove
outlawed_words = ['KG', ' KG', 'Tubers',' Tubers', 'Bunch', ' Bunch', 'pcs', ' pcs']
```

```python
# Convert date
df['date'] = pd.to_datetime(df['date'], dayfirst=True, format='mixed')
```

```python
# Engineer year and month
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
# Redenomination year condition
condition_1 = df['year'] == 2007

# Redenomination month condition
condition_2 = df['month'] <= 6

# Extract undesired period
redenomination = df[(condition_1) & (condition_2)]

# Drop undesired period
df.drop(redenomination.index, inplace=True)
```

```python
# Create an array of crops from the dataframe
unique_crop = list(df['commodity'].unique())
```

```python
chosen_crops = ['Cassava',
                    'Cowpeas',
                    'Cowpeas (white)',
                    'Eggplants',
                    'Maize',
                    'Maize (yellow)',
                    'Millet',
                    'Onions',
                    'Peppers (dried)',
                    'Peppers (fresh)',
                    'Rice (local)',
                    'Rice (paddy)',
                    'Sorghum',
                    'Soybeans',
                    'Tomatoes (local)',
                    'Tomatoes (navrongo)',
                    'Yam',
                    'Yam (puna)']
```

```python
df = df.drop(columns=['market_id', 
                      'latitude', 
                      'longitude', 
                      'commodity_id', 
                      'category',
                      'priceflag', 
                      'pricetype', 
                      'currency', 
                      'usdprice',
                      'Unnamed: 0'])
```

```python
# Standardize spelling region
df['region'] = df['region'].apply(lambda x: x.title())
```

```python
df.head()
```

```python
# Sort according to date
df = df.sort_values('date')
```

```python
# Split data frame into smaller data frames according to crops
for region in region_files:

    for unique in unique_crop:
    
        if unique in chosen_crops:
    
            # Retrieve original data frame and subset crop
            unique_df = df[df['commodity'] == unique]
            # unique_df = unique_df.drop(
                # ['crop_category', 'crop', 'crop_id'], axis=1)
    
            # Retrieve outlawed words and iterate
            for outlaw in outlawed_words:
    
                strings_to_remove = ['KG', ' KG', 'Tubers', ' Tubers', 'Bunch', ' Bunch', 'pcs', ' pcs']
    
                df['unit'] = df['unit'].apply(lambda x: str(x).replace(next((s for s in strings_to_remove if s in str(x)), ''), '').strip() or 1 if pd.notna(x) else 1)
                
                # if unique_df['unit'].str.contains(outlaw, na=False).any():
                #     unique_df['unit'] = unique_df['unit'].str.replace(
                #         outlaw, '')
                # else:
                #     continue
    
            # df['unit'] = df['unit'].fillna(1)
            
            # Convert unit to float
            df['unit'] = df['unit'].astype(dtype='int')
            
            # Unit conversion
            df['cedi_price/KG'] = df['price']/df['unit']
    
            # Drop final columns
            # df = df.drop(columns = ['year', 'month', 'unit', 'commodity'])
            
            # Naming convention
            unique = unique.replace(' ', '').lower()
            region = os.path.basename(region).lower().replace('.csv', '')
            delimiter = '_'
            unique_region_name = region+delimiter+unique
    
            # Define the directory path
            directory = 'data/crop_splits'
            
            # Check if the directory does not exist
            if not os.path.exists(directory):
                os.makedirs(directory)
            
            # Save split data frame to folder
            unique_df.to_csv(f'{directory}/{unique_region_name}.csv', index=False)
```
